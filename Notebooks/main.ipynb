### Normalization

- We will now normalize column titles, setting everything to lowercase.
- We will also be swapping blank spaces (" ") by underscores ("_").

```python
def col_lowercase(x):
    cols = []
    for item in x:
        cols.append(item.lower())
    return cols
```

```python
data.columns = col_lowercase(data.columns)
data.columns
```

```python
def no_spaces(x):
    cols = []
    for item in x:
        col = ""
        for letter in item:
            if letter == " ":
                letter = "_"
                col += letter
            else:
                col += letter
        cols.append(col)
    return cols
```

```python
data.columns = no_spaces(data.columns)
data.columns
```
